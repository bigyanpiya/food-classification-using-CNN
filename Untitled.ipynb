```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
import cv2
import numpy as np
from numpy import save
```

```python
class_names = ["banana", "apple", "pear", "grapes", "orange", "kiwi", "watermelon", "pomegranate", "pineapple", "mango", "cucumber", "carrot", "capsicum", "onion", "potato", "lemon", "tomato", "raddish", "beetroot", "cabbage", "lettuce", "spinach", "soy beans", "cauliflower", "bell pepper", "chilli pepper", "turnip", "corn", "sweetcorn", "sweetpotato", "paprika", "jalepeno", "ginger", "garlic", "peas", "eggplant"]

file_names = []
train_data_array = []
train_data_labels_array = []

print("Loading the training data")
```

```python
count = 0
for dirname, _, filenames in os.walk('Downloads/fv classification/train'):
    #print(dirname)
    for filename in filenames:
        frame = cv2.imread(os.path.join(dirname, filename))
        #print(os.path.join(dirname, filename))
        # Check the validity of the image
        if frame is None:
            print("not an image")
        else:
            resized = cv2.resize(frame, (28, 28), interpolation=cv2.INTER_AREA)
            checkSize = resized.shape[0]
            if checkSize == 28:
                file_names.append(dirname+"/"+filename)
                train_data_array.append(resized)
                index = class_names.index(os.path.basename(dirname))
                train_data_labels_array.append(index)
                #print(os.path.basename(dirname))
                count += 1
                print(count)
                print(dirname, filename)
```

```python
train_data = np.array(train_data_array)
train_data_labels = np.array(train_data_labels_array)

print("Finished loading the training data")
print("Number of training records : ", train_data.shape[0])

print(train_data.shape)
print(train_data_labels.shape)
```

```python
from matplotlib import pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```python
demoImage = train_data[900]
plt.imshow(cv2.cvtColor(demoImage, cv2.COLOR_BGR2RGB), interpolation='nearest')
plt.show()
index = train_data_labels[900]
print(class_names[index])
print(file_names[900])
```

```python
test_file_names = []
test_data_array = []
test_data_big_array = []
test_data_labels_array = []

print("Loading the testing data")
```

```python
count = 0
for dirname, _, filenames in os.walk('Downloads/fv classification/test'):
    #print(dirname)
    for filename in filenames:
        frame = cv2.imread(os.path.join(dirname, filename))
        #print(os.path.join(dirname, filename))
        # Check the validity of the image
        if frame is None:
            print("not an image")
        else:
            resizedBig = cv2.resize(frame, (280, 280), interpolation=cv2.INTER_AREA)
            resized = cv2.resize(frame, (28, 28), interpolation=cv2.INTER_AREA)
            checkSize = resized.shape[0]
            if checkSize == 28:
                test_file_names.append(dirname+"/"+filename)
                test_data_array.append(resized)
                test_data_big_array.append(resizedBig)
                index = class_names.index(os.path.basename(dirname))
                test_data_labels_array.append(index)
                #print(os.path.basename(dirname))
                count += 1
                print(count)
                print(dirname, filename)
```

```python
test_data = np.array(test_data_array)
test_data_big = np.array(test_data_big_array)
test_data_labels = np.array(test_data_labels_array)

print("Finished loading the testing data")
print("Number of testing records : ", test_data.shape[0])

print(test_data.shape)
print(test_data_labels.shape)
```

```python
train_data = train_data / 255.0
test_data = test_data / 255.0
```

```python
model = keras.Sequential([
    # first we will flatten the images. We will take the 28x28x3 and flatten the shape as the input to the model
    keras.layers.Flatten(input_shape=(28,28,3)), # this is the input layer
    
    # let's define the hidden layer
    # we don't know what is the exact number, so we will try with 512 neurons
    keras.layers.Dense(512, activation='relu'), # relu has no negative values
    
    # this is the last layer - the classification of the images
    # we have 36 Classes (Apple, Banana, Orange, .....)
    keras.layers.Dense(36, activation='softmax') # softmax has return values between 0 to 1
])

print("Finished building the model skeleton")
```

```python

# compile the model
model.compile(
    # optimizer -> calculate the gradient descent of the network
    optimizer='adam',
    # loss function
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy'] # metrics measurement
)

print("Finished compile the model")
```

```python
# train the model
# starting with 120 epochs, will tweak it accordingly to increase the efficiency
model.fit(train_data, train_data_labels, epochs=120)

print("Finished training the model")
# Accuracy: 0.9450 Loss: 0.2161
```

```python
test_loss, test_acc = model.evaluate(test_data, test_data_labels, verbose=1) # verbose is a parameter of how detailed is the log in the console
print("************ Test accuracy : ", test_acc)

# Test Results | Accuracy: 0.955 ## very good result
```

```python
searchIndex = 40
predictions = model.predict(test_data)

predict_index = np.argmax(predictions[searchIndex])
#print(class_index)

predict_name = class_names[predict_index]

demoImage = test_data_big[searchIndex]
#print(demoImage)
plt.imshow(demoImage, interpolation='nearest')
plt.show()
real_index = test_data_labels[searchIndex]
print("Real Class : ", class_names[real_index]) # Showing the real name
print("Predict Class : ", class_names[predict_index]) # Showing the predicted name
print(test_file_names[searchIndex])
```

```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import preprocess_input

def preprocess_new_image(image_path):
    # Load the image from the file
    img = image.load_img(image_path, target_size=(28,28))  # Set the target height and width

    # Convert the image to a NumPy array
    img_array = image.img_to_array(img)

    # Expand dimensions to match the input shape of your model (e.g., (1, height, width, channels))
    #img_array = np.expand_dims(img_array, axis=0)

    # Preprocess the image using the same preprocessing used during training
    preprocessed_image = preprocess_input(img_array)

    return preprocessed_image
```

```python
from tensorflow.keras.models import load_model


# Preprocess the new image (replace 'new_image.jpg' with the path to your new image)
new_image = preprocess_new_image('Downloads/or.jpg')

# Predict the class for the new image
predictions = model.predict(np.array([new_image]))

# Get the predicted class index
predict_index = np.argmax(predictions)

# Get the predicted class name
predict_name = class_names[predict_index]

# Display the new image
rescaled_image = np.clip(new_image, 0, 255).astype(np.uint8)
plt.imshow(rescaled_image, interpolation='nearest')
plt.show()



# Print the predicted class
print("Predicted Class: ", predict_name)
```
